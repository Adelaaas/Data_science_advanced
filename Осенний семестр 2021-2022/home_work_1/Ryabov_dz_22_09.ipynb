```python
a, b = input().split()
a, b = int(a), int(b)
for i in range(a):
    c = list(input())
    fl = 0
    try:
        for j in range(len(c)):
            if c[j+b] != c[j]:
                if c[j] != '?' and c[j+b] != '?':
                    fl += 1
                    break
                else:
                    c[j+b] = c[j]
    except Exception:
        pass
    if fl == 0:
        print("YES")
    else:
        print('NO')
```

```python
from sklearn.preprocessing import StandardScaler
from catboost import CatBoostClassifier
import pandas as pd

scaler = StandardScaler()
```

```python
test = pd.read_csv("C:\\Users\In9.var\Desktop\Projects\data1\\test.csv")
test
```

```python
train_target = pd.read_csv("C:\\Users\In9.var\Desktop\Projects\data1\\train_target.csv")
train_target
```

```python
transactions_test = pd.read_csv("C:\\Users\In9.var\Desktop\Projects\data1\\transactions_test.csv")
transactions_test
```

```python
transactions_train = pd.read_csv("C:\\Users\In9.var\Desktop\Projects\data1\\transactions_train.csv")
transactions_train
```

```python
agg_features=transactions_train.groupby('client_id')['amount_rur'].agg(['sum','mean','std','min','max']).reset_index()
counter_df_train=transactions_train.groupby(['client_id','small_group'])['amount_rur'].sum()
cat_counts_train=counter_df_train.reset_index().pivot(index='client_id', \
                                                      columns='small_group',values='amount_rur')
cat_counts_train=cat_counts_train.fillna(0)
cat_counts_train.columns=['small_group_'+str(i) for i in cat_counts_train.columns]
cat_counts_train.head()
```

```python
train=pd.merge(train_target,agg_features,on='client_id')
train=pd.merge(train,cat_counts_train.reset_index(),on='client_id')

agg_features_test=transactions_test.groupby('client_id')['amount_rur'].agg(['sum','mean','std','min','max']).reset_index()
counter_df_test=transactions_test.groupby(['client_id','small_group'])['amount_rur'].sum()
cat_counts_test=counter_df_test.reset_index().pivot(index='client_id', columns='small_group',values='amount_rur')
cat_counts_test=cat_counts_test.fillna(0)
cat_counts_test.columns=['small_group_'+str(i) for i in cat_counts_test.columns]
```

```python
test=pd.merge(test,agg_features_test,on='client_id')
test=pd.merge(test,cat_counts_test.reset_index(),on='client_id')
```

```python
y_train=train['bins']
X_train=train.drop(['client_id', 'bins'], axis=1)
X_test=test.drop(['client_id'], axis=1)
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
model = CatBoostClassifier()
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
pred
```
