К стоп-словам, как правило, относят предлоги, союзы, междометия, частицы и некоторые части речи, которые часто встречаются в тексте, являются служебными и не  большой смысловой нагрузки – являются избыточными

Системе необходимо игнорировать такие слова при анализе

Способы поулчения:
*   Создать свой список
*   Взять готовый список из библиотек (к примеру nltk)


```python
from nltk.corpus import stopwords

en_stpw = stopwords.words("english")

en_stpw
```

```python
from nltk.corpus import stopwords
ru_stpw = stopwords.words("russian")

ru_stpw
```

Расширение списка стоп слов

```python
ru_stpw.extend(['ы', 'нею'])

ru_stpw[-5:]
```

Удаление стоп слов из текста

```python
text = 'once the cat went for a walk and well he saw such a beautiful tree'

fw = [word for word in text.split() if word not in stopwords.words('english')]
nfw = [word for word in text.split() if word in stopwords.words('english')]
print(nfw, '- stop words')
print(fw, '- remaining words')
print(' '.join(fw), '- the resulting sentence')
```
