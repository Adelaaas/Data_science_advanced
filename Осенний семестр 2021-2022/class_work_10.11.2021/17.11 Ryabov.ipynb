```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd


lang_into = ['da', 'bg', 'cs', 'de', 'el', 'es', 'et', 'fi',
     'fr', 'hu', 'it', 'lt', 'lv', 'nl', 'pl', 'pt',
     'ro', 'sk', 'sl', 'sv']

lang = {'train_da':set(), 'train_bg':set(), 'train_cs':set(), 'train_de':set(), 'train_el':set(), 'train_es':set(), 'train_et':set(),
        'train_fi':set(), 'train_fr':set(), 'train_hu':set(), 'train_it':set(), 'train_lt':set(), 'train_lv':set(), 'train_nl':set(),
        'train_pl':set(), 'train_pt':set(), 'train_ro':set(), 'train_sk':set(), 'train_sl':set(), 'train_sv':set()}

path = "C:\\Users\In9.var\Desktop\Projects\data2\\"
df = pd.DataFrame(columns=['word', 'language'])
k=0
for j in lang.keys():
    tfidf = TfidfVectorizer(encoding='utf-8',
                        ngram_range=(1, 2),
                        stop_words=None,
                        lowercase=True,
                        max_features=300,
                        norm='l2',
                        analyzer="char",
                        sublinear_tf=True)
    f = open(path+j, "r", encoding='utf-8')
    l = tfidf.fit_transform(f).toarray()
    feature_names = tfidf.get_feature_names()
    for i in feature_names:
        if '.' in i or ',' in i or '?' in i or '-' in i or ';' in i or ':' in i or i.split() == [] or "\"" in i or i.isdigit():
            pass
        else:
            df.loc[k] = [i, j[-2:]]
            k+=1
df
```

```python
def func(word):
    lis = list(word)
    for i in range(len(word[:-1])):
        lis.append(word[i]+word[i+1])
    return lis
```

```python
ans = []
b = 0
with open(path+"test", "r", encoding='utf-8') as f:
    for i in f.readlines():
        l = {}
        i = i[2:-1].lower()
        for j in func(i):
            p = list(df['word'][df['word'] == j].index)
            for h in p:
                if df.loc[h][1] not in l.keys():
                    l[df.loc[h][1]] = 0
                l[df.loc[h][1]] += 1
        ans.append(max(l.keys(), key=lambda x: l[x]))
        print(ans[-1])
print(ans)
```

```python
otvet = pd.DataFrame(columns=['lang_samp'])
for i in range(len(ans)):
    otvet.loc[i] = ans[i]
otvet.to_csv('C:\\Users\In9.var\Desktop\Projects\data2\lang_submition.csv')
```
