```python
from sklearn.linear_model import LogisticRegression # логистическая регрессия
from sklearn.tree import DecisionTreeClassifier # дерево решений
from sklearn.svm import SVC # метод опорных векторов
import xgboost as xgb

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(
                                                train.drop(columns=['critical_temperature']),
                                                train['critical_temperature'],
                                                test_size=1/3,
                                                random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
answers_pred = model.predict(X_train)

print(model.score(X_test, y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X_train, y_train)

answers_pred = model.predict(X_train)

print(model.score(X_test, y_test))
```

```python
from google.colab import files
answers_pred = model.predict(test)
answer = pd.DataFrame({'Index': [i for i in range(len(answers_pred))],'critical_temp': answers_pred})
answer
```
