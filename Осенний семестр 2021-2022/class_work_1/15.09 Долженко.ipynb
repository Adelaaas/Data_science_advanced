```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score

from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data Science/11 КЛАСС/Уроки/dataset.csv')
df
```

```python
print(df.info(), "\n___________________________________________________________\n")
print(df.isna().sum())
```

```python
df = df.drop(['name', 'prefix','diameter', 'albedo', 'diameter_sigma',
              'id', 'spkid','full_name', 'pdes',
              'orbit_id', 'epoch','epoch_mjd', 'epoch_cal', 'equinox'], axis = 1)
```

```python
df = df.dropna(how='any')
```

```python
print(df.info(), "\n____________________________________________________\n")
print(df.isna().sum())
```

```python
df = pd.get_dummies(df, columns=['class', 'neo'])

x = LabelEncoder()
l = x.fit_transform(df['pha'])
df['pha'] = l

df.info()
```

```python
X = df.drop('pha', axis = 1)
y = df['pha']

standart = StandardScaler()
X_st = standart.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_st, y, test_size = 1/3, random_state = 42)
```

```python
lg = LogisticRegression()
lg.fit(X_train, y_train)
model_predict = lg.predict(X_test)
print("\nACCURACY:", accuracy_score(y_test, model_predict), "\nRECALL:", recall_score(y_test, model_predict), "\nPRECISION:", precision_score(y_test, model_predict) )
```

```python
tree = DecisionTreeClassifier(random_state=1)
tree.fit(X_train, y_train)
model_predict = tree.predict(X_test)
print("\nACCURACY:", accuracy_score(y_test, model_predict), "\nRECALL:", recall_score(y_test, model_predict), "\nPRECISION:", precision_score(y_test, model_predict) )
```

```python
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
model_predict = xgb.predict(X_test)
print("\nACCURACY:", accuracy_score(y_test, model_predict), "\nRECALL:", recall_score(y_test, model_predict), "\nPRECISION:", precision_score(y_test, model_predict) )
```

```python
svm = SVC()
svm.fit(X_train, y_train)
model_predict = svm.predict(X_test)
print("\nACCURACY\tRECALL\tPRECISION\n", 
      accuracy_score(y_test, model_predict), "\t", recall_score(y_test, model_predict), "\t", precision_score(y_test, model_predict), "\n")
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
model_predict = knn.predict(X_test)
print("\nACCURACY:", accuracy_score(y_test, model_predict), "\nRECALL:", recall_score(y_test, model_predict), "\nPRECISION:", precision_score(y_test, model_predict) )
```

В итоге результаты моделей XGBClassifier и DecisionTreeClassifier показали почти идеальные результаты.
Модель LogisticRegression при увеличении количества итераций (не дефолтные) ощутимо увеличивает время работы в разы. Время работы KNeighborsClassifier и SVC также слишком велико
