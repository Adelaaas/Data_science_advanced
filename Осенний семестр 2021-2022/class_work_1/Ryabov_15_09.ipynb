```python
import pandas as pd


df = pd.read_csv("C:\\Users\In9.var\Desktop\Projects\data\\dataset.csv")
df
```

```python
df.info()
```

```python
df = df.drop(['id', 'spkid', 'full_name', 'name', 'prefix', 'equinox', 'orbit_id', 'pdes'], axis=1)
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df['pha'].value_counts()
```

```python
df['pha'] = df['pha'].map({'Y': 1, 'N': 0})
df = pd.get_dummies(df)
df
```

```python
X = df.drop('pha', axis=1)
y = df['pha']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_st = scaler.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
from sklearn.neighbors import KNeighborsClassifier as knn
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
models = ['CatBoostClassifier()', 'svm.SVC()', 'LogisticRegression()', 'knn(5)', 'RandomForestClassifier(max_depth=2, random_state=0)']
for i in range(len(models)):
    model = eval(models[i])
    model.fit(X_train, y_train)
    models[i] = model.score(X_test, y_test)
```

все модели показали прекрасные результаты, но наилучшим(по версии .score) признан RandomForestClassifier
