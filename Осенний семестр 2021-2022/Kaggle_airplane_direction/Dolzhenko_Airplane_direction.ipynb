```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report

from sklearn.tree import DecisionTreeClassifier
from catboost import CatBoostClassifier
from xgboost.sklearn import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import  KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sklearn.model_selection import train_test_split
```

```python
X = train.drop(columns=['target'], axis= 1)
y = train['target']

norm = MinMaxScaler()
X_norm = norm.fit_transform(X)
```

```python
Xn_train, Xn_test, y_train, y_test = train_test_split(X_norm, y, test_size=1/3, random_state=42)
```

```python
models = [DecisionTreeClassifier(), CatBoostClassifier(), XGBClassifier(), LogisticRegression(), KNeighborsClassifier(), RandomForestClassifier(), SVC()]
models_names = ['DecisionTreeClassifier', 'CatBoostClassifier', 'XGBClassifier', 'LogisticRegression', 'KNeighborsClassifier', 'RandomForestClassifier', 'SVC']
```

```python
for i in range(7):
    model = models[i]
    model.fit(Xn_train, y_train)
    prediction = model.predict(Xn_test)
    print(models_names[i], '\n----------------------------------------------------------------------\n', 
          classification_report(y_test, prediction), '\n----------------------------------------------------------------------\n')
```
