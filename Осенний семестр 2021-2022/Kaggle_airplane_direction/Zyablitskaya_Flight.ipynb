```python
import pandas as pd
test=pd.read_csv("/content/drive/MyDrive/Анализ Данных/Kaggle Flight/Data/test_airplanes.csv")
train=pd.read_csv("/content/drive/MyDrive/Анализ Данных/Kaggle Flight/Data/train_airplanes.csv")
train.head()
```

```python
from sklearn.linear_model import LogisticRegression # логистическая регрессия
from sklearn.tree import DecisionTreeClassifier # дерево решений
from sklearn.svm import SVC # метод опорных векторов
import xgboost as xgb

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix

X_train, X_test, y_train, y_test = train_test_split(
                                                train.drop(columns=['index','target','Unnamed: 0']),
                                                train['target'],
                                                test_size=1/3,
                                                random_state=42)
```

```python
log = LogisticRegression()
tree = DecisionTreeClassifier()
svc = SVC()
xbg = xgb.XGBClassifier()

models = [log, tree, svc, xbg]

for model in models:
    #  fit() - обучение модели/классификатора
    model.fit(X_train, y_train)

    # predict - тестирование
    y_pred = model.predict(X_test)

    # оценка классификатор/модели
    score = accuracy_score(y_test, y_pred)
    print(model)
    print(score)
    print('-'*50)
```

```python
model = xgb.XGBClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

score = accuracy_score(y_test, y_pred)
score
```
