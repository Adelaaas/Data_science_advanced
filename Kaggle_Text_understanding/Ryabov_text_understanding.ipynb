```python
import pandas as pd
import json

def get_X_y(data_json_file):
    X, y = [], []
    with open(data_json_file, 'r', encoding="utf-8") as json_file:
        json_list = list(json_file)
        #print(json_list[0])
        for json_str in json_list:
            item = json.loads(json_str)
            text = item['passage']['text']
            #print(item['passage'].keys())
            questions = item['passage']['questions']
            for q in questions:
                query = q['question']
                ans = q['answers']
                for a in ans:
                    X.append(text+' Query: '+query+' Answer: '+a['text'])
                    y.append(a['label'])
    return X, y

X_train, y_train = get_X_y('C:\\Users\In9.var\Desktop\Projects\data1\\train.jsonl')

train_df = pd.DataFrame({
    'text': X_train,
    'label': y_train
})
train_df
```

```python
train_df["text"] = train_df["text"].str.replace("(", "")
train_df["text"] = train_df["text"].str.replace(")", "")
train_df["text"] = train_df["text"].str.replace(".", "")
train_df["text"] = train_df["text"].str.replace("«", "")
train_df["text"] = train_df["text"].str.replace("»", "")
train_df["text"] = train_df["text"].str.replace("—", "")
train_df["text"] = train_df["text"].str.replace("  ", " ")
train_df.iloc[100]["text"]
```

```python
def get_X_test(data_json_file):
    X1 = []
    X2 = []
    X3 = []
    with open(data_json_file, 'r', encoding="utf-8") as json_file:
        json_list = list(json_file)
        for json_str in json_list:
            item = json.loads(json_str)
            text = item['passage']['text']
            questions = item['passage']['questions']
            for q in questions:
                query = q['question']
                ans = q['answers']
                for a in ans:
                    X1.append(text) #текст
                    X2.append(query) #вопрос
                    X3.append(a['text']) #ответ на вопрос
    return X1, X2, X3

X1, X2, X3 = get_X_test('C:\\Users\In9.var\Desktop\Projects\data1\\val_wtihout.jsonl')

test_df = pd.DataFrame({
    'text': X1,
    'query': X2,
    'answer': X3
})

test_df
```

```python
for n in range(20):
    test_df["text"] = test_df["text"].str.replace(f"({n})", "")
test_df["text"] = test_df["text"].str.replace(".", "")
test_df["text"] = test_df["text"].str.replace(",", "")
test_df["text"] = test_df["text"].str.replace("«", "")
test_df["text"] = test_df["text"].str.replace("»", "")
test_df["text"] = test_df["text"].str.replace("—", "")
test_df["text"] = test_df["text"].str.replace("  ", " ")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_df['text'], train_df['label'], test_size=0.33, random_state=42)

vectorizer = CountVectorizer(ngram_range=(2,5), analyzer='word', max_features = 10000)

X_train_vector = vectorizer.fit_transform(X_train).toarray()
X_test_vector = vectorizer.transform(X_test).toarray()
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier,LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import classification_report

rs = 42

clf = LogisticRegression(random_state=rs)   
clf2 = RandomForestClassifier(random_state=rs, n_jobs =-1)
#clf4 = SVC(random_state=rs)
clf5 = DecisionTreeClassifier(random_state=rs)

clflist = [clf, clf2, clf5]

for clf in clflist:
    print(clf)
    clf.fit(X_train_vector, y_train)
    predictions = clf.predict(X_test_vector)
    print(classification_report(y_test, predictions))
    print('='*50)
```
