```python
corpus = [
          'Text of first document.',
          'Text of the second document made longer.',
          'Number three.',
          'This is number four.',
]

list_of_words = []
x = []
for i in range(0, len(corpus), 1):
  for j in range(0, len(corpus[i].split()), 1):
    list_of_words.append(corpus[i].split()[j]) 

print (list_of_words)



amount = 0
for i in range(0, len(corpus), 1):
  amount = len(corpus[i].split()) + amount

print(amount)

```

```python
from collections import Counter
list_text =[]
for i in range(0, len(list_of_words), 1):
  list_text = list_of_words[i].lower().replace('.', '').split() + list_text

count = Counter(list_text)
print(count)
```
