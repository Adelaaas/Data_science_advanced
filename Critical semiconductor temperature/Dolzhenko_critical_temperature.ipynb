```python
# X = train.drop(columns=['critical_temp', 'material'])
X = train.drop(columns=['critical_temperature'])
y = train['critical_temperature']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=48)
```

```python
from catboost import CatBoostRegressor
```

```python
ctb = CatBoostRegressor()
ctb.fit(X_train, y_train)
# ctv_pred = ctb.predict(X_test)
print(ctb.score(X_test, y_test))
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
print(lr.score(X_test, y_test))
```

```python
tree = DecisionTreeRegressor()
tree.fit(X_train, y_train)
print(tree.score(X_test, y_test))
```

```python
rnf = RandomForestRegressor()
rnf.fit(X_train, y_train)
print(rnf.score(X_test, y_test))
```

```python
xgb = XGBRegressor()
xgb.fit(X_train, y_train)
print(xgb.score(X_test, y_test))
```
