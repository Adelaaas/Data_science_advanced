```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# в качестве исследования проверим, с какими данными результат получится лучше. для этого нормализуем и стандартизируем их
scaler = StandardScaler()
minmax = MinMaxScaler()
```

```python
from sklearn.metrics import classification_report, mean_absolute_error
from sklearn.model_selection import KFold, cross_val_score

from sklearn.tree import DecisionTreeRegressor
from catboost import CatBoostRegressor
from xgboost.sklearn import XGBRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import  KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import train_test_split
```

```python
# исслудуем какая модель и с какими данными покажет наилучшие результаты
models = [CatBoostRegressor(), DecisionTreeRegressor(), XGBRegressor(), LinearRegression(), KNeighborsRegressor(), RandomForestRegressor()]
models_names = ['CatBoostRegressor', 'DecisionTreeClassifier', 'XGBRegressor', 'LinearRegression', 'KNeighborsRegressor', 'RandomForestRegressor']
```

```python
X = train.drop(['Predict','Unnamed: 0'], axis = 1)
y = train['Predict']
```

```python
X_norm = minmax.fit_transform(X)
X_st = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 43)
```

```python
# обычные данные

for i in range(len(models)):
    model = models[i]
    model_name = models_names[i]

    model.fit(X_train, y_train)
    predicted = model.predict(X_test)

    print('\n____________________', model_name, '_______________________\n')
    # print(classification_report(y_test, predicted))
    print("MAE: {:3f}".format(mean_absolute_error(y_test, predicted)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size=0.2, random_state = 43)
```

```python
# нормализованные данные

for i in range(len(models)):
    model = models[i]
    model_name = models_names[i]

    model.fit(X_train, y_train)
    predicted = model.predict(X_test)

    print('\n____________________', model_name, '_______________________\n')
    # print(classification_report(y_test, predicted))
    print("MAE: {:3f}".format(mean_absolute_error(y_test, predicted)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_st, y, test_size=0.2, random_state = 43)
```

```python
# стандартизированные данные

for i in range(len(models)):
    model = models[i]
    model_name = models_names[i]

    model.fit(X_train, y_train)
    predicted = model.predict(X_test)

    print('\n____________________', model_name, '_______________________\n')
    # print(classification_report(y_test, predicted))
    print("MAE: {:3f}".format(mean_absolute_error(y_test, predicted)))
```

```python
# LinearRegression показал наилучший результат
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 43)
```

```python
# {'colsample_bytree': 0.8, 'eta': 0.375, 'gamma': 0.9, 'max_depth': 11, 'min_child_weight': 1.0, 'n_estimators': 887, 'subsample': 0.7000000000000001}
model = LinearRegression()
model.fit(X_train, y_train)
pred = model.predict(X_test)
print("MAE: {:3f}".format(mean_absolute_error(y_test, pred)))
```
