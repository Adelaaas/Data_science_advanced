обработка пустых значений


```python
mean_class = train['MinTemp'].mean()
print(round(mean_class))

train['MinTemp'] = train['MinTemp'].fillna(mean_class)
print(train.isna().sum())
```

```python
mean_class = train['MaxTemp'].mean()
print(round(mean_class))

train['MaxTemp'] = train['MaxTemp'].fillna(mean_class)
print(train.isna().sum())
```

```python
mean_class = train['Rainfall'].mean()
print(round(mean_class))

train['Rainfall'] = train['Rainfall'].fillna(mean_class)
print(train.isna().sum())
```

```python
mean_class = train['Evaporation'].mean()
train['Evaporation'] = train['Evaporation'].fillna(mean_class)

mean_class = train['Sunshine'].mean()
train['Sunshine'] = train['Sunshine'].fillna(mean_class)

mean_class = train['WindGustSpeed'].mean()
train['WindGustSpeed'] = train['WindGustSpeed'].fillna(mean_class)

mean_class = train['WindSpeed9am'].mean()
train['WindSpeed9am'] = train['WindSpeed9am'].fillna(mean_class)

mean_class = train['WindSpeed3pm'].mean()
train['WindSpeed3pm'] = train['WindSpeed3pm'].fillna(mean_class)

mean_class = train['Humidity9am'].mean()
train['Humidity9am'] = train['Humidity9am'].fillna(mean_class)

mean_class = train['Humidity3pm'].mean()
train['Humidity3pm'] = train['Humidity3pm'].fillna(mean_class)

mean_class = train['Pressure9am'].mean()
train['Pressure9am'] = train['Pressure9am'].fillna(mean_class)

mean_class = train['Pressure3pm'].mean()
train['Pressure3pm'] = train['Pressure3pm'].fillna(mean_class)

mean_class = train['Cloud9am'].mean()
train['Cloud9am'] = train['Cloud9am'].fillna(mean_class)

mean_class = train['Cloud3pm'].mean()
train['Cloud3pm'] = train['Cloud3pm'].fillna(mean_class)

mean_class = train['Temp9am'].mean()
train['Temp9am'] = train['Temp9am'].fillna(mean_class)

mean_class = train['Temp3pm'].mean()
train['Temp3pm'] = train['Temp3pm'].fillna(mean_class)
print(train.isna().sum())
```

```python
train = train.dropna()
print(train.isna().sum())
```

преобразовываем категориальные признаки

```python
train = pd.get_dummies(train, columns=['Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X = train.drop(['RainTomorrow'], axis=1)
y = train['RainTomorrow']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)

model = LogisticRegression()
model.fit(X_train, y_train)
answers_pred = model.predict(X_train)
print(model.score(X_test, y_test))
```

```python
otvet = pd.DataFrame(answers_pred, columns=['RainTomorrow'])
otvet.reset_index(inplace=True)

otvet
```
