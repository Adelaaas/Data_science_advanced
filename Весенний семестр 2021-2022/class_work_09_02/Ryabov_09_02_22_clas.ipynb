```python
df = df.drop(["Location", "WindDir9am", "WindDir3pm", "WindGustDir"], axis=1)
df = df.fillna(0)
X = df.drop(["RainTomorrow"], axis=1)
y = df["RainTomorrow"]
```

Протестируем модели

```python
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
from sklearn.neighbors import KNeighborsClassifier as knn
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
models = ['CatBoostClassifier()', 'svm.SVC()', 'LogisticRegression()', 'knn(5)', 'RandomForestClassifier(max_depth=2, random_state=0)']
for i in range(len(models)):
    model = eval(models[i])
    model.fit(X_train, y_train)
    models[i] = model.score(X_test, y_test)
```
