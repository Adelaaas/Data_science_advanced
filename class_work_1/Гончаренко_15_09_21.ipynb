```python
df['pha']=df['pha'].map({'Y' : 1, 'N' : 0})
df=pd.get_dummies(df)
df
```

```python
X = df.drop('pha', axis=1)
y = df['pha']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)	#Делим данные на тстовые и тренировочные
scaller = StandardScaler()
X_train = scaller.fit_transform(X_train)
X_test = scaller.transform(X_test)
```

```python
models = [1,2,3,4,5]
l = LogisticRegression()
l.fit(X_train, y_train)
models[0]=l.score(X_test, y_test)

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
models[1]=knn.score(X_test, y_test)

sv = SVC()
sv.fit(X_train, y_train)
models[2]=sv.score(X_test, y_test)

rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
models[3]=rfc.score(X_test, y_test)

tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
models[4]=tree.score(X_test, y_test)
for model in models:
  print(model)
#дерево имеет наилучшую точность и показало наилучший результат.
```
