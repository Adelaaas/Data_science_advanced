```python
**Машинное обучение** - это 
 обширный подраздел искусственного интеллекта, изучающий методы построения алгоритмов, способных обучаться.
**Виды машинного обучения:**
1. Обучение без учителя - это один из способов машинного обучения, при котором испытуемая система спонтанно обучается выполнять поставленную задачу без вмешательства со стороны экспериментатора.
2. Обучение с учителем - это обучение предполагает наличие полного набора размеченных данных для тренировки модели на всех этапах ее построения.

**Обучение с учителем:**
- Задачи регрессии   - это когда по заданному набору признаков необходимо спрогнозировать некую целевую переменную.
- Задачи классификации - это задача разделения множества наблюдений (объектов) на группы, называемые классами, на основе анализа их формального описания

**Обучение без учителя:**
- Кластеризация - это то еще один алгоритм обучения без присмотра, который используется для группировки непомеченных точек данных, имеющих сходные характеристики. 
```

```python
Методы оценки моделей:
MSE - MSE расшифровывается как «Mean Squared Error» и переводится как «Средняя квадратическая ошибка». 
Суть метода заключается по сути в том, чтобы минимизировать сумму квадратов отклонений фактических значений
 от расчётных (SSE - «Sum of Squared Errors»). Если полученную сумму разделить на число наблюдений, то получится та самая MSE. (для регрессии)
 R-squered -коэффициент детерминации. Характеризует степень сходства исходных данных и предсказанных. В отличии от MSE не зависит от единиц измерения данных, поэтому поддается сравнению.
 Accuracy - деление всхе правильных ответов на все вообще ответы
 Recall - В этой метрике мы рассматриваем только P-случаи: когда в реальных наблюдениях было "да". И считаем, какую долю из этих случаев модель предсказала правильно.(для классификации)
 Precision - Какая часть наших предсказаний "да" действительно сбылась
```

```python
df = df.drop(['name', 'prefix'], axis=1) #удалила ненужные колонки
df = df.drop(['id', 'spkid', 'full_name', 'equinox', 'orbit_id', 'pdes'], axis=1)
df.isnull().sum() #кол-во NaN объектов в каждой колонке
```

```python
df = df.dropna() #удаляет все строки, где есть значение NaN
```

```python
df['pha'] = df['pha'].map({'Y': 1,'N': 0}) #rename Y на 1 и N на 0
df = pd.get_dummies(df)
df
```

```python
X = df.drop('pha', axis=1) #таблица без pha
y = df['pha'] #cстолбец pha
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) #разбиваем на тестовые и обучение данные + перемешиваем данные traintrstsplit

# cross_val_score
scaler = StandardScaler() #стандартизирует, удаляет минимизирует сильновыделяющиеся значения
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC #метод опорных векторов
from sklearn.ensemble import RandomForestClassifier



model = [DecisionTreeClassifier(), LogisticRegression(),  RandomForestClassifier()]
for i in range(0, 3, 1):
  model[i].fit(X_train, y_train)
  model[i].score(X_test, y_test)

```
